```python
import random

import numpy as np
import pandas as pd

from NeuralNet import ActivationType, NeuralNetwork

```

```python
def one_hot_encode(y: np.ndarray, num_classes: int) -> np.ndarray:
	num_samples: int = y.shape[1]
	y_encoded = np.zeros((num_classes, num_samples))

	for i, label in enumerate(y[0, :]):
		y_encoded[label, i] = 1.0

	return y_encoded

```

```python
train_df = pd.read_csv("./data/mnist_train.csv")
test_df = pd.read_csv("./data/mnist_test.csv")

x_train = (
	train_df.iloc[:, 1:].transpose().values
)  # shape = 784 * m, so each col is a sample
x_train = x_train / 255.0  # to squish the pixel values between 0-1 instead of 0-255
y_train = one_hot_encode(train_df.iloc[:, 0:1].values.reshape((1, -1)), 10)  # shape = 10 * m


x_test = test_df.iloc[:, 1:].values.transpose()
x_test = x_test / 255.0
y_test = one_hot_encode(test_df.iloc[:, 0:1].values.reshape((1, -1)), 10)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python

np.random.seed(42)
nn = NeuralNetwork(
	structure=[784, 24, 24, 10],
	activation_types=[ActivationType.Relu, ActivationType.Relu, ActivationType.Sigmoid],
)

nn.print_stat(x_test, y_test)
```

```python
nn.train(
	x_train,
	y_train,
	number_of_epochs=100,
	batch_size=128,
	learning_rate=0.1,
	constant_lr=False,
	decay_rate=0.05,
	verbose=True,
)

nn.print_stat(x_train, y_train)
nn.print_stat(x_test, y_test)
```
